# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=df, orient='v')
plt.title('Boxplot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

plt.figure(figsize=(12, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*2, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query = """
SELECT DatabaseDesireNextYear, COUNT(*) AS count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5;
"""

# Execute the query and store the results in a DataFrame
df = pd.read_sql_query(query, conn)

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'],autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Query to get the median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
query = """
SELECT Age,
       AVG(WorkWeekHrs) AS MedianWorkWeekHrs,
       AVG(CodeRevHrs) AS MedianCodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
ORDER BY Age;
"""

# Execute the query and store the results in a DataFrame
df= pd.read_sql_query(query, conn)

# Create a stacked bar chart
df.plot(kind='bar', x='Age', stacked=True,
                    y=['MedianWorkWeekHrs', 'MedianCodeRevHrs'],
                    figsize=(10, 6), width=0.8)

plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.xticks(rotation=0)
plt.legend(['Median Work Week Hours', 'Median Code Review Hours'])
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# Query to get the median ConvertedComp for all ages from 45 to 60
query = """
SELECT Age,
       AVG(ConvertedComp) AS MedianConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
ORDER BY Age;
"""

# Execute the query and store the results in a DataFrame
age_comp_data = pd.read_sql_query(query, conn)

# Plot the median ConvertedComp for all ages from 45 to 60
plt.figure(figsize=(12, 6))
plt.plot(age_comp_data['Age'], age_comp_data['MedianConvertedComp'], marker='o')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Query to get the count of each MainBranch
query = """
SELECT MainBranch, COUNT(*) AS count
FROM master
GROUP BY MainBranch
ORDER BY count DESC;
"""

# Execute the query and store the results in a DataFrame
main_branch_data = pd.read_sql_query(query, conn)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(main_branch_data['MainBranch'], main_branch_data['count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Count of Respondents by Main Branch')
plt.gca().invert_yaxis()  # Invert y-axis to have the largest bar on top
plt.show()

```
